```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import confusion_matrix
```

```python
weekly = pd.read_csv("Weekly.csv")
```

```python
weekly.head()
```

```python
weekly.shape
```

```python
import seaborn as sns
sns.pairplot(weekly)
plt.show()
```

Volume vs Today looks looks interesting... the points with higher volume has a higher variance in Today. Lets take a closer look at that. 

```python
from sklearn import datasets, linear_model
Vol = weekly.as_matrix(["Volume"])
Today = weekly.as_matrix(["Today"])
lr = linear_model.LinearRegression()
lr.fit(Vol,Today)
plt.scatter(Vol, Today,  color='black')
plt.plot(Vol, lr.predict(Vol), color='blue',
         linewidth=3)
plt.show()
```

Doesn't tell us much... lets look at the difference between up days and down days with volume. 

```python
def f(row):
    if row['Direction'] == 'Up':
        return 1
    return 0
weekly["Direction2"] = weekly.apply(f, axis = 1)
X = weekly.as_matrix(["Volume","Direction2"])
Y = weekly.as_matrix(["Today"])
lr = linear_model.LinearRegression()
lr.fit(X,Y)
print (lr.coef_, lr.intercept_)
```

```python
weekly_up = weekly[weekly.Direction2==1]
weekly_down = weekly[weekly.Direction2==0]
X_up = weekly_up.as_matrix(["Volume"])
Y_up = weekly_up.as_matrix(["Today"])
X_down = weekly_down.as_matrix(["Volume"])
Y_down = weekly_down.as_matrix(["Today"])
lr.fit(X_up, Y_up)
plt.scatter(X_up, Y_up,  color='black')
plt.plot(X_up, lr.predict(X_up), color='black',
         linewidth=2)
lr.fit(X_down, Y_down)
plt.scatter(X_down, Y_down,  color='blue')
plt.plot(X_down, lr.predict(X_down), color='blue',
         linewidth=2)
sns.despine()
plt.show()
```

From this, we can draw a reasonable conclusion that a lot of movement in the stock(either up or down) drives a higher volume. This makes intuitive sense; lots of people will sell to collect profits/buy for momentum if the stock moves up a lot. In addition, lots of people will sell to minimize loss/buy b/c the share is cheaper if the stock moves down a lot.

## Confusion Matrix

```python
results = logReg.predict(testing)
print(results.shape)
results
```

```python
print(confusion_matrix(weekly["Direction"], results))
```

Figuring out which row corresponds to "Up" and which row corresponds to "Down" - 

```python
up_correct = 0
up_predicted_as_down = 0
for num in range(1089):
    if (weekly["Direction"][num]=="Up" and results[num]=="Up"):
        up_correct+=1
    if (weekly["Direction"][num]=="Up" and results[num]=="Down"):
        up_predicted_as_down +=1
print(up_correct)
print(up_predicted_as_down)
```

Interpretation of the confusion matrix - 
55 Downs and 558 Ups are predicted correctly. 47 Ups are predicted as downs and 429 Downs are predicted as ups.

Overall fraction of correct predictions - (55+558)/ (47+429+55+558) = 0.563

## Chapter 5 Question 8 

```python
import random
random.seed(1)
X = np.random.randn(100)
y = np.random.randn(100)
y = X - (2 *X**2) + np.random.randn(100) 
```

```python
plt.scatter(X,y)
plt.show()
```

We notice the Y has a much higher variance than X, since we squared each value of X to get Y. It additionally looks like a parabola due to squaring X. 

```python
# LOOCV
import sklearn.linear_model as skl_lm
from sklearn.cross_validation import train_test_split, LeaveOneOut, KFold, cross_val_score

random.seed()
def f(row):
    return row['X'] * row['X']
df = pd.DataFrame({'X': list(X),'Y':list(y)}, index = None)

df["X2"] = df['X'] **2
df["X3"] = df['X'] **3
df["X4"] = df['X'] **4
df.head()
    
```

```python
X = df.as_matrix(["X"])
Y = df.as_matrix(["Y"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
score = cross_val_score(lr, X, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X: ", score)
```

```python
X2 = df.as_matrix(["X2"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
features = df[list({'X','X2'})].values
score = cross_val_score(lr, features, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X, X^2: ", score)
```

```python
X3 = df.as_matrix(["X3"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
features = df[list({'X','X2','X3'})].values
score = cross_val_score(lr, features, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X, X^2, X^3: ", score)
```

```python
X4 = df.as_matrix(["X4"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
features = df[list({'X','X2','X3','X4'})].values
score = cross_val_score(lr, features, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X, X^2, X^3, X^4: ", score)
```

The best score(smallest error)  came from the model with X and X^2. This makes intuitive sense since Y is a parabola. 

```python
##Repeat with different random seed 
import sklearn.linear_model as skl_lm
from sklearn.cross_validation import train_test_split, LeaveOneOut, KFold, cross_val_score

random.seed(20)
X = np.random.randn(100)
y = np.random.randn(100)
y = X - (2 *X**2) + np.random.randn(100) 
df = pd.DataFrame({'X': list(X),'Y':list(y)}, index = None)

df["X2"] = df['X'] **2
df["X3"] = df['X'] **3
df["X4"] = df['X'] **4

X = df.as_matrix(["X"])
Y = df.as_matrix(["Y"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
score = cross_val_score(lr, X, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X: ", score)

X2 = df.as_matrix(["X2"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
features = df[list({'X','X2'})].values
score = cross_val_score(lr, features, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X, X^2: ", score)

X3 = df.as_matrix(["X3"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
features = df[list({'X','X2','X3'})].values
score = cross_val_score(lr, features, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X, X^2, X^3: ", score)

X4 = df.as_matrix(["X4"])
lr = linear_model.LinearRegression()
loo = LeaveOneOut(df.index.size)
features = df[list({'X','X2','X3','X4'})].values
score = cross_val_score(lr, features, Y, cv=loo, scoring='mean_squared_error').mean()
print("LooCV with X, X^2, X^3, X^4: ", score)
```

The fit with X,X^2 is still the best b/c it is still a parabola. Obviously, the exact LooCV errors are different since there are new randomized points.  The coeffecients show how much each column is weighted when predicting Y. It would make sense for X^2 to have the largest coeffecient since the data is quadratic. 

# Problem 9 

```python
beta_hats = []
for _ in range(100):
    x = np.random.randn(51,50)
    e = np.random.randn(51,1) *.5
    b = np.zeros(shape = (50,1))
    b.fill(1)
    y = np.dot(x,b) + e
    lr = linear_model.LinearRegression()
    lr.fit(x,y)
    y = lr.predict(x)
    beta_hats.append(lr.coef_)
print("Mean of beta hat: ", np.mean(beta_hats))
print("Variance of single component of beta hat: ", np.var(beta_hats[0]))
```

```python
from sklearn.linear_model import LassoCV
import warnings
warnings.filterwarnings('ignore')
beta_hats = []
for _ in range(100):
    x = np.random.randn(51,50)
    e = np.random.randn(51,1) *.5
    b = np.zeros(shape = (50,1))
    b.fill(1)
    y = np.dot(x,b) + e
    model_lasso = LassoCV(alphas = [0.01, 0.1, 1, 10, 100]).fit(x, y)
    beta_hats.append(model_lasso.coef_)
print("Mean of beta hat: ", np.mean(beta_hats))
print("Variance of single component of beta hat: ", np.var(beta_hats[0]))
```

We notice that the mean and variance are similar to the original beta hats of the first linear regression.
